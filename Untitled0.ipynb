```python
#load overall medal count
medals_total=pd.read_csv("/medals_total.csv")
medals_total.head()
```

```python
#load Coaches Data
Coaches=pd.read_excel("/Coaches.xlsx")
Coaches

```

```python
#load medal winner dataset
event_medal=pd.read_csv("/medals.csv")
event_medal.head()

```


# Data Cleaning

1. Overall Medals

```python
# Cleaning medal_totals by removing unnecessary columns

medals_total = medals_total.drop('Country Code', axis=1) # drop the Country Code column
medals_total.head()
```

```python
# sort out data to include those countries that have at least 10 gold medals
medal_ten = medals_total[medals_total['Gold Medal'] >= 10] # Filter rows based on condition
medal_ten
```

Below are bar graphs that indicate the amount of medals obtained by each of the countries that obtained at least 10 gold medals.  Note: ROC=Republic of China or Taiwan

```python

fig1a = px.histogram(medal_ten, x="Country", y = "Gold Medal", color = "Country", barmode="group", labels={"Country": "Country","Gold Medal": "Gold Medal"})
fig1a.update_layout(title="Gold Medal Count by Country")
fig1a.show()
```

```python
fig1b = px.histogram(medal_ten, x="Country", y = "Silver Medal", color = "Country", barmode="group", labels={"Country": "Country","Silver Medal": "Silver Medal"})
fig1b.update_layout(title="Silver Medal Count by Country")
fig1b.show()
```

```python
fig1c = px.histogram(medal_ten, x="Country", y = "Bronze Medal", color = "Country", barmode="group", labels={"Country": "Country","Bronze Medal": "Bronze Medal"})
fig1c.update_layout(title="Bronze Medal Count by Country")
fig1c.show()
```

```python
fig1d = px.histogram(medal_ten, x="Country", y = "Total", color = "Country", barmode="group", labels={"Country": "Country","Total": "Total"})
fig1d.update_layout(title="Total Medal Count by Country")
fig1d.show()
```

From this one could conclude that most of the countries that had obtained 10 gold medals have at least a large degree of investment in Olympic success and a good number of candidates go to perform at the Olympics.

```python
#For future reference for use in the other data sets
countries = medal_ten['Country'].to_numpy() # create array from the countries in medal_ten
countries
```

2. Participating Athletes

```python
#Sorted out the athletes list to those who participate for the 10 nations who scored 10 gold medals during the 2021 Olympics
Athletes_sorted = Athletes[Athletes['NOC'].isin(countries)].sort_values('NOC') # Filter and sort the DataFrame
Athletes_sorted # Display the first few rows of the sorted DataFrame
```

```python
#overall athlete count by the 10 gold medal array
count = Athletes_sorted['NOC'].value_counts()
count
```

The largest number of athletes come from the USA with second coming from the host country Japan and Australia.  The People's Republic of China is fourth despite having a large population.  All entries exceed 200 participants

```python
#count of the selected athletes by discipline
count = Athletes_sorted['Discipline'].value_counts()
count
```

The largest number of candidates are in the athletics section (Track and Field) followed by swimming and rowing.

```python
#Sort out the number of participants by country and discipline and get a head count of them

noc_discipline_counts = Athletes_sorted.groupby(['NOC', 'Discipline'])['Name'].count().reset_index(name='Count')
noc_discipline_counts_sorted = noc_discipline_counts.sort_values(['NOC', 'Count'], ascending=[True, False])
print(noc_discipline_counts_sorted)
```

```python
#bar graph of number of athletes in each discipline divided by each of the countries with at least 10 gold medals
fig_noc_discipline = px.bar(noc_discipline_counts_sorted,
                             x='Discipline',
                             y='Count',
                             color='NOC',
                             title='Number of Athletes per Discipline by Country',
                             labels={'NOC': 'Country', 'Count': 'Number of Athletes'})
fig_noc_discipline.show()
```

It looks like the most participants are in Athletics with the USA taking a good number of spots to participate.  It is more noticed in Athletics, Rowing, and Swimming  In other events such as Cycling Track, Diving, and Table Tennis, the apparent differences diminish.

3. Coaches

```python
#sorted the coaches table by country and discipline and drop the Event column
Coaches.sort_values(by = ['NOC', 'Discipline']) # Sort the dataframe by country and discipline
Coaches_cleaned = Coaches.drop('Event', axis=1)
Coaches_cleaned
```

```python
# Using the countries array, kept coaches who represented those countries
Coaches_cleaned[Coaches_cleaned.NOC.isin(countries)] # Filter the dataframe to include coaches in the specified countries
```

```python
coach_discipline_counts = Coaches_cleaned.groupby(['NOC', 'Discipline'])['Name'].count().reset_index(name='Count')
coach_discipline_counts_sorted = noc_discipline_counts.sort_values(['NOC', 'Count'], ascending=[True, False])
print(coach_discipline_counts_sorted)
```

```python
fig_coach_discipline= px.bar(coach_discipline_counts_sorted,
                             x='Discipline',
                             y='Count',
                             color='NOC',
                             title='Number of Coaches per Discipline by Country',
                             labels={'NOC': 'Country', 'Count': 'Number of Coaches'})
fig_coach_discipline.show()
```

There is a similar trend in coaches to athletes in terms of distribution of resources to ensure a success in the countries with more than 10 gold medals to ensure success.

4. Medal winning events during 2020-2021 Olympics

```python
#Cleaned/sorted the data set for medal winners and dropped unnecessary columns
# Drop specified columns
event_medal = event_medal.drop(['medal_type', 'medal_date', 'athlete_short_name', 'athlete_link', 'country_code', 'discipline_code'], axis=1)

# Sort the DataFrame
event_medal_sorted = event_medal.sort_values(by=['discipline', 'country'])

# Display the cleaned and sorted DataFrame
event_medal_sorted
```

```python
 # Split the DataFrame into two based on the 'country' column
top_countries = ['United States of America', "People's Republic of China", 'Japan',
       'Great Britain', 'ROC', 'Australia', 'Netherlands', 'France',
       'Germany', 'Italy']
top_countries_medal = event_medal_sorted[event_medal_sorted['country'].isin(top_countries)]
other_countries_medal = event_medal_sorted[~event_medal_sorted['country'].isin(top_countries)]

# Display the resulting DataFrames
print("Top Countries Medal:")
print(top_countries_medal)

print("\nOther Countries Medal:")
print(other_countries_medal)
```
